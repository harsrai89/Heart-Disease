Splitting data into train and test sets.

```python
array = df.values
X = array[:, 0:13]
y = array[:, 13]

seed = 7
tsize = 0.2
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=tsize, random_state=seed)
```

**logistic**

```python
from sklearn.linear_model import LogisticRegression
# instantiate the model (using the default parameters)
logreg = LogisticRegression()

# fit the model with data
logreg.fit(X_test,y_test)
```

```python
X_train[1]
```

```python
logreg.predict([[49. ,   1. ,   2. , 118. , 149. ,   0. ,   0. , 126. ,   0. ,
         0.8,   2. ,   3. ,   2. ]])
```

```python
accuracy2 = logreg.score(X_test,y_test)
print("accuracy = ",accuracy2 * 100,"%")
```

```python
import matplotlib.pyplot as plt
import numpy as np
def plot_confusion_matrix(cm, title='Confusion matrix', cmap=plt.cm.Blues):
    target_names=['yes','no']
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(target_names))
    plt.xticks(tick_marks, target_names, rotation=45)
    plt.yticks(tick_marks, target_names)
    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

```python
from sklearn.metrics import confusion_matrix
y_pred=logreg.predict(X_test)
cm=confusion_matrix(y_test, y_pred)
print('Confusion matrix, without normalization')
print(cm)
plot_confusion_matrix(cm)
```

**KNN**

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=12)
knn.fit(X_test, y_test)
```

```python
print(knn.predict([[49. ,   1. ,   2. , 118. , 149. ,   0. ,   0. , 126. ,   0. ,
         0.8,   2. ,   3. ,   2. ]]))
```

```python
accuracy4= knn.score(X_test,y_test)
print("accuracy = ",accuracy4* 100,"%")
```

```python
from sklearn.metrics import confusion_matrix
y_pred=knn.predict(X_test)
cm=confusion_matrix(y_test, y_pred)
print('Confusion matrix, without normalization')
print(cm)
plot_confusion_matrix(cm)
```

```python
ALGOS=["LOGISTIC","KNN"]
ACC=[accuracy2,accuracy4]
plt.bar(ALGOS,ACC)
plt.xlabel("Algorithms")
plt.ylabel("Accuracy")
plt.title("Comparison")
```
